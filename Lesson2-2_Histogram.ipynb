```python
import numpy as np, matplotlib.pyplot as plt
```

```python
samples=100000
x0=np.random.uniform(0,1,samples)
x1=np.random.normal(0,1,samples)
x2=np.random.triangular(0,3,5,samples)
x3=np.random.exponential(2,samples)
x4=np.random.chisquare(20,samples)
x5=np.random.power(3,samples)

x = [x0,x1,x2,x3,x4,x5]
```

```python
def make_da_plot(bins=None):
    fig = plt.figure(1,figsize=(14,14/1.25))
    labels = ["Uniform", "Normal", "Tri", "Exp", r"$\chi^2$", "Pow"]
    colors = "red blue orange black green purple".split(' ')

    Naxis=321

    for i in range(len(x)):
        ax=fig.add_subplot(Naxis)
        ax.hist(x[i], color=colors[i], bins=bins)
        ax.set_title(labels[i])
        Naxis += 1
    
    fig.suptitle("A collection of random data from several distributions.", fontsize=16)
    #fig.show() # avoid warnings about the matplotlib backend 
    
make_da_plot()
```

```python
make_da_plot(50)
```

```python
make_da_plot(250)
```

```python
make_da_plot(np.sqrt(samples).astype(int))
```
