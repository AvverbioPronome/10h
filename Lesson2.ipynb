```python
with open("square_roots.tsv", 'r') as fp:
    n, r = list(), list()
    for ln in fp:
        x, y = map(lambda x: float(x), ln.split("\t"))
        plt.plot(x,y, '.')
        n.append(x)
        r.append(y)
```

```python
a=np.array(list(zip(n,r)))
```

```python
def misure_sfera(r):
    return {
        "cerchio_massimo": 2 * np.pi * r,
        "superficie": 4 * np.pi * r**2,
        "volume": 4/3 * np.pi * r**3}
```

```python
misure_sfera(3.215)
```

```python
n = np.linspace(0,10,1000)
plt.plot(n, misure_sfera(n)['volume'], '-')
```

```python
xx = np.arange(1,100) # Avoid complaints about trying to calculate Log(0)
yy = np.sqrt(xx)

fig = plt.figure()
ax = fig.add_axes([.15, .15, .8, .8])
ax.minorticks_on()
ax.plot(xx, yy, label="Sqrt")
ax.plot(xx, np.log2(xx), '.r', label="Log2")
ax.set_xlabel("x-data", fontsize = 20)
ax.set_ylabel("y-data", fontsize = 12)
ax.set_title("Some title.title", fontsize=22)
ax.text(5, 1, "Some text.text", fontsize=20)
ax.legend()

ax2 = fig.add_axes([1,.15,.8,.8])

xxx = np.linspace(-10*np.pi, 10*np.pi, 1000)
ax2.plot(xxx, np.sin(xxx)/xxx, 'g-', lw=.8, label=r'$ \frac{\sin(x)}{x} $')
ax2.text(7, 0.25, r'$\frac{\sin{x}}{x}$', fontsize=40)
ax2.legend(fontsize=20)
ax2.grid()
ax2.set_title("Some function--function", fontsize=22)
```
