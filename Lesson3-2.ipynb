```python
import numpy as np, matplotlib.pyplot as plt
```

```python
plt.rc("figure", figsize=(10,7))
```

```python
x = np.linspace(-3, 20, 100)
x = np.random.normal(loc=x, scale=0.25, size=len(x))

y = 1.2 * x**2 - 0.3*x + 2.8
y = np.random.normal(loc=y, scale=0.25, size=len(y))
```

```python
plt.plot(x,y,'.r')
```

```python
cc, cov = np.polyfit(x, y, deg=2, cov=True)  # this fits. cov is a covariance matrix

xx = np.linspace(-3,21) 
yy = np.polyval(cc, xx) # this evaluates polynomes: cc is a list of coefficients.

plt.plot(xx,yy,'b-')
```

```python
np.sqrt(cov.diagonal())
```

```python
xp = [1,2,3]
fp = [3,2,0]

np.interp(2.5, xp, fp)
np.interp([0, 1, 1.5, 2.72, 3.14], xp, fp) #che dovrebbe fare sta cosa?
plt.plot(xp,fp,'o')
np.all(np.diff(xp)>0)
```

```python
x = np.linspace(0, 2*np.pi,10)
y = np.sin(x)

xvals = np.linspace(0, 2*np.pi, 500)
yinterp = np.interp(xvals, x, y)

plt.plot(x, y, 'o', xvals, yinterp, '-')
```

```python
from scipy.interpolate import interp2d
```

```python
x = y = np.arange(-5.01, 5.01, .25)
xx, yy = np.meshgrid(x, y)
z = np.sin(xx**2+yy**2)
f = interp2d(x,y,z, kind='cubic')
```

```python
xnew = ynew = np.arange(-5.01, 5.01, 1e-2)
znew = f(xnew, ynew)
plt.figure()
plt.pcolormesh(xx,yy,z)
plt.figure()
plt.pcolormesh(xnew,ynew,znew)
```

```python
x = np.linspace(0,4,13)
y = np.array([0,2,3,3.5,3.75,3.875,3.9375,4])
X, Y = np.meshgrid(x,y)
Z = np.sin(np.pi*X/2) * np.exp(Y/2)
plt.figure()
plt.pcolormesh(X,Y,Z)

x2 = y2 = np.linspace(0,4,65)
X2, Y2 = np.meshgrid(x2,y2)
f = interp2d(x, y, Z, kind='cubic')
Z2=f(x2,y2)

plt.figure()
plt.pcolormesh(x2,y2,Z2)
```
