```python
import numpy as np, matplotlib.pyplot as plt
from scipy import integrate
```

```python
plt.rc("figure", figsize=(9,6))
```

```python
def fun1(x):
    return np.exp(-x**2)
def fun2(x, a, p, s):
    return a * np.exp(-(p-x)**2/s**2)
```

```python
x= np.linspace(-4,8,1000)
a, p, s = 1.5, 1, 1.5
plt.plot(x, fun1(x), x, fun2(x, a, p, s))
```

```python
I1 = integrate.quad(fun1, -np.inf, np.inf)
print (I1, I1[0] -np.sqrt(np.pi))
```

```python
I2 = integrate.quad(fun2, -2, 5, args=(a,p,s))
print (I2)
```
