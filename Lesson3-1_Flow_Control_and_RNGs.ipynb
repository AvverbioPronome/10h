```python
mu = 5
s = np.random.poisson(mu, 10000)

plt.hist(s, density=True, bins=10)
```

```python
mu = 0
sigma = .5
from scipy.stats import norm
s = np.random.normal(mu, sigma, 10000)

plt.hist(s, density=True, bins=100)
x = np.linspace(-4*sigma,4*sigma,1000)
plt.plot(x,norm.pdf(x, mu, sigma), lw=5, alpha=.8)
plt.show()
```
