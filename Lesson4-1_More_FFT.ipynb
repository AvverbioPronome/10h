```python
import numpy as np, matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq
```

```python
plt.rc("figure", figsize=(15,6))
```

```python
f = 400
ns = 16
T = 1/f
dt = T/ns

def fun(t,f):
    return np.sin(2*np.pi*f*t) - 1/2*np.sin(2*np.pi*2*f*t) - 1/2*np.sin(2*np.pi*f/4*t)
```

```python
t, step=np.linspace(0, dt*(ns*3.25-1), int(ns*3.25), retstep=True)
y=np.pad(fun(t,f), (0, max([0,64-len(t)])), mode='constant', constant_values=0)
t=np.arange(0, dt*(ns*8-1), step)[0:len(y)] # LE MAZZATE.

ln = len(y)
Y3 = fft(y)/ln
F3 = fftfreq(ln,dt)

I = F3>=0
fig = plt.figure()
axf = fig.add_subplot(121)
axf.plot(t,y,'k-')
axf.plot(t,y,'ro')

axfft = fig.add_subplot(122)
axfft.vlines(F3[I], 0, np.abs(Y3[I]))
axfft.set_ylim(0, np.abs(Y3[I]).max())
axfft.set_xlabel("Frequency [Hz]")
axfft.set_ylabel("Amplitude")
```
