```python
import numpy as np, matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
plt.rc("figure", figsize=(15, 10))
```

```python
def dy_dx(y, x):
    return x-y

xs = np.linspace(0,5,5)
x = np.linspace(0,5,1000)

y0=1.0
ys = odeint(dy_dx, y0, xs)

plt.plot(xs, ys, 'o', x, x-1+2*np.exp(-x), '-')
```

```python
def dydt(y, t):
    a = -2
    b = -.1
    return np.array([y[1], a * y[0] + b * y[1]])

time = np.linspace(0,10,100)
yinit = [.0005, .2]
y = odeint(dydt, yinit, time)
ax = plt.plot(time, y)
plt.legend(ax, ['y', 'y\''])
```

```python
def dydt(y, t):
    return [y[1], -2*y[0] - y[1]]

t=np.arange(0,15, .01)
y=odeint(dydt, [1,0], t)
ax=plt.plot(t,y)
plt.legend(ax, ['y', 'y\''])
```

```python
def solva(Y, t):
    return [Y[1], -2*Y[0] - Y[1]]

t = np.linspace(-10,10,1000)
y = odeint(solva, [0,1], t)
plt.plot(t,y)
```
