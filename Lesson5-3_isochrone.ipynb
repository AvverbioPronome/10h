```python
import numpy as np, matplotlib.pyplot as plt
import os

directory = "course_materials/isocz303y251p00o0d0e0/FEHm070/"
isochrones = dict()
for file in sorted(os.listdir(directory)):
    isochrones[file] = np.loadtxt(os.path.join(directory,file))
```

```python
fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(111)

for key in isochrones:
    isochrone = isochrones[key]
    ax.plot(isochrone[:,1], isochrone[:,0], lw=1, label=key[:5])

plt.legend()
fig.gca().invert_xaxis()
ax.set_xlabel(r"$T_e$", fontsize=20)
ax.set_ylabel(r"$\frac{L}{L_{\odot}}$", fontsize=20)
ax.set_ylabel(r"${L}/{L_{\odot}}$", fontsize=20)
```
